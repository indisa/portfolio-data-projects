#  Исследование результатов А/В-теста и поиск инсайтов

SollmaFin — международное мобильное приложение для онлайн-торговли финансовыми активами. Пользователи могут инвестировать деньги в акции, валюту, криптовалюту, биржевые фонды (ETF) и другие активы. Целевая аудитория приложения — начинающие инвесторы. Получить доступ к финансовым рынкам можно со смартфона либо через веб-версию. 

Приложение ориентировано на рынок Латинской Америки с акцентом на четыре страны: Мексику, Бразилию, Колумбию и Аргентину — там быстро растёт интерес к финтеху и инвестициям, а экономики очень разные. 

Появилась гипотеза о недостаточной финансовой грамотности пользователей: они не понимают, как работать с активами разной степени риска. Клиенты покупают активы с высоким риском и теряют деньги, что снижает вовлечённость.
Команда продукта решила обновить онбординг пользователей и добавить в него детальную информацию о различиях финансовых активов и связанных с ними рисках. Однако появилось опасение, что углублённый онбординг отпугнёт клиентов от пополнения депозита и более рискованных вложений. 

Чтобы разобраться, команда решила провести A/B-эксперимент. 

### А/В-эксперимент

В эксперименте участвовали новые пользователи, которые зарегистрировались в приложении со 2 по 15 июня 2025 года. 
Пользователей случайным образом разделили на две равные группы: 
- Контрольная группа проходила стандартный онбординг без обязательного обучения.
- Тестовая группа проходила обновлённый онбординг с подробной информацией об активах и связанных с ними рисках. 

После разделения активность пользователей анализировали в течение недели.

Цель эксперимента — оценить, как обновлённый онбординг влияет на поведение пользователей и их дальнейшую инвестиционную активность.

При внедрении новой фичи команда продукта выдвинула такие гипотезы:
- Гипотеза роста: обучающий онбординг помогает пользователям лучше понимать принципы инвестирования, поэтому они будут чаще открывать второй депозит.
- Гипотеза риска: информация о возможных потерях и высоких рисках отпугнёт некоторых новичков, особенно самых осторожных, что снизит конверсию в первый депозит.
- Дополнительная гипотеза: после нового онбординга пользователи, которые выбрали высокорискованные активы, будут чаще, чем раньше, возвращаться и открывать второй депозит. При старом онбординге пользователи часто покупали активы с высоким риском без понимания последствий. Это приводило к потерям и оттоку после первого депозита. 

Чтобы увидеть полную картину, недостаточно следить за метриками конверсии. Команда выбрала для эксперимента такой набор метрик: 
- Ключевая метрика — средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию).
- Барьерная метрика — конверсия из регистрации в первый депозит.
- Вспомогательная метрика 1 — конверсия из первого депозита во второй.
- Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

Ожидалось, что ключевая и барьерная метрики не упадут, а вспомогательные покажут значительный рост.

**Задача 2.2.** Анализ воронок событий.
- Постройте классическую и пошаговую воронки;
- Охарактеризуйте пользовательский путь в приложении и определите узкие места воронки. 


````python
# Задаём последовательность этапов нашего приложения
steps = [
    'install / open_web',
    'introduction',
    'registration',
    'main_page',
    'onboarding_complete',
    'first_deposit',
    'second_deposit'
]

# Считаем уникальных пользователей на каждом шаге
counts = {}
for step in steps:
    counts[step] = (
        hist[hist['event_name'] == step]
        ['user_id']
        .nunique()
    )

# Общая конверсия от первого шага
first_count = counts[steps[0]]
pct_from_start = {step: counts[step] / first_count * 100 for step in steps}

# Пошаговая конверсия от предыдущего шага
pct_from_prev = {steps[0]: 100.0}
for i in range(1, len(steps)):
    prev = counts[steps[i-1]]
    curr = counts[steps[i]]
    pct_from_prev[steps[i]] = curr / prev * 100 if prev > 0 else 0

# Собираем дф для визуализации
funnel_df = pd.DataFrame({
    'step':          steps,
    'users':         [counts[s] for s in steps],
    'pct_from_start': [pct_from_start[s] for s in steps],
    'pct_from_prev':  [pct_from_prev[s]  for s in steps],
})

display(funnel_df)

# Рисуем классическую воронку
fig1 = px.funnel(
    funnel_df,
    x='pct_from_start',
    y='step',
    title='Классическая воронка (% от первого шага)',
    labels={'pct_from_start': '% от старта', 'step': 'Этап'}
)
fig1.show()

# Рисуем пошаговую воронку
fig2 = px.funnel(
    funnel_df,
    x='pct_from_prev',
    y='step',
    title='Step-by-step воронка (% от предыдущего шага)',
    labels={'pct_from_prev': '% от предыдущего шага', 'step': 'Этап'}
)
fig2.show()
````

````python
def build_funnel(df: pd.DataFrame, segment_name: str) -> pd.DataFrame:
    records = []
    for seg_value, sub in df.groupby(segment_name):
        counts = {step: sub.loc[sub['event_name']==step, 'user_id'].nunique() for step in steps}
        first = counts[steps[0]]
        pct_start = [counts[s]/first*100 for s in steps]
        pct_prev  = []
        for i, s in enumerate(steps):
            if i == 0:
                pct_prev.append(100.0)
            else:
                prev = counts[steps[i-1]]
                pct_prev.append((counts[s]/prev*100) if prev>0 else 0)
        for step, cnt, ps, pp in zip(steps, counts.values(), pct_start, pct_prev):
            records.append({
                segment_name: seg_value,
                'step': step,
                'users': cnt,
                'pct_from_start': ps,
                'pct_from_prev': pp
            })
    return pd.DataFrame(records)

# Собираем воронку для платформ
funnel_plat = build_funnel(hist, 'platform')

# Классическая воронка (от старта)
fig1 = px.funnel(
    funnel_plat,
    x='pct_from_start',
    y='step',
    color='platform',
    title='Классическая воронка по платформам',
    labels={'pct_from_start':'% от старта','step':'Этап'}
)
fig1.show()

# Пошаговая воронка (от предыдущего)
fig2 = px.funnel(
    funnel_plat,
    x='pct_from_prev',
    y='step',
    color='platform',
    title='Step-by-step воронка по платформам',
    labels={'pct_from_prev':'% от предыдущего','step':'Этап'}
)
fig2.show()
````

**Задача 2.3.** Влияние уровня риска актива на открытие второго депозита.
- Изучите, как уровень риска первого актива влияет на конверсию во второй депозит.  

````python
# Берём первый купленный актив (asset_purchase) и его уровень риска
asset_first = (
    hist[hist['event_name']=='asset_purchase']
    .sort_values('event_ts')
    .drop_duplicates(subset='user_id', keep='first')
    .loc[:, ['user_id','risk_level']]
    .dropna(subset=['risk_level'])  # убираем тех, у кого нет risk_level
)

# Список тех, кто сделал первый и тех, кто сделал второй депозит
first_dep = (
    hist[hist['event_name']=='first_deposit']
    .drop_duplicates(subset='user_id')
    .loc[:, ['user_id']]
)
second_dep = (
    hist[hist['event_name']=='second_deposit']
    .drop_duplicates(subset='user_id')
    .assign(second=1)
    .loc[:, ['user_id','second']]
)

# Объединяем всё в один DataFrame
conv = (
    asset_first
      .merge(first_dep,  on='user_id', how='inner')  # только у тех, кто сделал asset_purchase и first_deposit
      .merge(second_dep, on='user_id', how='left')   # помечаем, дошёл ли до второго
      .assign(second=lambda df: df['second'].fillna(0))
)

# Считаем CR по уровням риска
cr_table = (
    conv
    .groupby('risk_level')
    .agg(
        users_with_first  = ('user_id', 'count'),
        users_with_second = ('second',  'sum')
    )
    .assign(
        cr_percent = lambda df: df['users_with_second'] / df['users_with_first'] * 100
    )
    .reset_index()
)

# Выводим таблицу и строим график
print("Таблица конверсии из первого во второй депозит по уровням риска:")
display(cr_table)

plt.figure(figsize=(6,4))
plt.bar(cr_table['risk_level'], cr_table['cr_percent'])
plt.title('Конверсия из 1-го во 2-й депозит по уровням риска')
plt.xlabel('Уровень риска актива')
plt.ylabel('CR, %')
plt.ylim(0, cr_table['cr_percent'].max() + 5)
plt.show()
````

**Промежуточные выводы**

Самая низкая повторная конверсия—у «high-risk» (26%), то есть только каждый четвёртый из тех, кто выбрал высокорисковые активы, возвращается на второй депозит. Наоборот, «low-risk» дают почти половину повторных депозитов (46%), а «medium» — чуть меньше (43%). Разрыв между крайними группами очень велик (20 п. п.), что говорит о том, что уровень риска актива сильно влияет на желание пользователя реинвестировать. Этот результат явно показывает необходимость фокусироваться на удержании после первого депозита, особенно в сегменте «high-risk».

````python
# Суммарные депозиты по каждому пользователю
user_deposits = (
    hist
    .groupby('user_id')['amount']
    .sum()
    .reset_index(name='total_deposit')
)

# Дата регистрации каждого пользователя
reg_dates = (
    hist[['user_id','first_dt']]
    .drop_duplicates(subset='user_id')
)

# Склейка, заполняем NaN (тех, кто не вносил депозит) нулём
user_reg = (
    reg_dates
    .merge(user_deposits, on='user_id', how='left')
    .fillna({'total_deposit': 0})
)

# Средняя сумма депозитов на пользователя по дате регистрации
daily_metric = (
    user_reg
    .groupby('first_dt')['total_deposit']
    .mean()
    .reset_index(name='avg_deposit')
)

# Таблица и график
display(daily_metric.head())

plt.figure(figsize=(10,4))
plt.plot(daily_metric['first_dt'], daily_metric['avg_deposit'], marker='o')
plt.title('Динамика ключевой метрики\n(средняя сумма депозитов на пользователя)')
plt.xlabel('Дата регистрации')
plt.ylabel('Средняя сумма депозитов')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

````python
# Делаем user-level таблицу
user_summary = (
    hist
    .pivot_table(
        index=['user_id','country_code','platform','first_dt'],
        columns='event_name',
        values='amount',
        aggfunc='sum',
        fill_value=0
    )
    .reset_index()
)
# Переименуем колонки
user_summary = user_summary.rename(columns={
    'first_deposit': 'sum_first_deposit',
    'second_deposit':'sum_second_deposit'
})

# Общая сумма всех депозитов на пользователя
user_summary['total_deposit'] = (
    user_summary['sum_first_deposit'] + user_summary['sum_second_deposit']
)

# Флаги конверсий
user_summary['made_first']  = (user_summary['sum_first_deposit']  > 0).astype(int)
user_summary['made_second'] = (user_summary['sum_second_deposit'] > 0).astype(int)

# Сводка по странам
seg_country = (
    user_summary
    .groupby('country_code')
    .agg(
        n_users        = ('user_id',       'nunique'),
        avg_deposit    = ('total_deposit','mean'),
        conv_first     = ('made_first',   'mean'),
        conv_second    = ('made_second',  'mean')
    )
    .reset_index()
)

# Сводка по платформам
seg_platform = (
    user_summary
    .groupby('platform')
    .agg(
        n_users        = ('user_id',       'nunique'),
        avg_deposit    = ('total_deposit','mean'),
        conv_first     = ('made_first',   'mean'),
        conv_second    = ('made_second',  'mean')
    )
    .reset_index()
)

display(seg_country, seg_platform)


# Средний депозит по странам
plt.figure(figsize=(6,3))
plt.bar(seg_country['country_code'], seg_country['avg_deposit'])
plt.title('Средний депозит на пользователя по странам')
plt.ylabel('Avg deposit')
plt.show()

# Конверсия в первый депозит по платформам
plt.figure(figsize=(6,3))
plt.bar(seg_platform['platform'], seg_platform['conv_first']*100)
plt.title('Конверсия в первый депозит по платформам')
plt.ylabel('% конверсии')
plt.show()
````

````python
# Считаем базовые метрики по полной выборке
user_deposits = hist.groupby('user_id')['amount'].sum()
mean_all = user_deposits.mean()
std_all = user_deposits.std()
p99 = user_deposits.quantile(0.99)

print(f"До фильтрации: mean = {mean_all:.2f}, std = {std_all:.2f}")
print(f"99-й перцентиль (порог выброса): {p99:.2f}")

# Очищаем хвост, оставляем только 99% пользователей
user_trim = user_deposits[user_deposits <= p99]

# Пересчитываем метрики на «обрезанной» выборке
mean_trim = user_trim.mean()
std_trim  = user_trim.std()

print(f"После фильтрации до 99%: mean = {mean_trim:.2f}, std = {std_trim:.2f}")
````

````python

# Задаём этапы воронки
stages = [
    'install / open_web',
    'introduction',
    'registration',
    'main_page',
    'onboarding_complete',
    'first_deposit',
    'second_deposit'
]

# Функция для подсчёта воронки по группе
def get_funnel(df, group_name):
    sub = df[df['group'] == group_name]
    # флаги, дошёл ли пользователь до этапа
    user_stage = (
        sub[['user_id','event_name'] ]
        .drop_duplicates()
        .pivot(index='user_id', columns='event_name', values='event_name')
        .notna()
        .astype(int)
        .reindex(columns=stages, fill_value=0)
    )
    counts = user_stage.sum()
    # накопительная конверсия от первого шага
    pct_from_start = counts / counts.iloc[0] * 100
    # пошаговая конверсия
    pct_from_prev = [100.]
    for i in range(1, len(stages)):
        prev = counts.iloc[i-1]
        curr = counts.iloc[i]
        pct_from_prev.append(curr / prev * 100 if prev else 0)
    return counts, pct_from_start, pd.Series(pct_from_prev, index=stages)

# Собираем воронки для control и test
counts_c, start_c, step_c = get_funnel(abt, 'control')
counts_t, start_t, step_t = get_funnel(abt, 'test')

funnel_df = pd.DataFrame({
    'stage': stages,
    'control_count': counts_c.values,
    'test_count': counts_t.values,
    'control_from_start': start_c.values,
    'test_from_start': start_t.values,
    'control_from_prev': step_c.values,
    'test_from_prev': step_t.values,
})

display(funnel_df)

# Классическая воронка (накопительная конверсия)
fig = go.Figure()
fig.add_trace(go.Funnel(
    name='control',
    y=stages,
    x=funnel_df['control_from_start'],
    textinfo="value+percent initial"
))
fig.add_trace(go.Funnel(
    name='test',
    y=stages,
    x=funnel_df['test_from_start'],
    textinfo="value+percent initial"
))
fig.update_layout(title="Классическая воронка (% от первого шага)")
fig.show()

# Step-by-step воронка (конверсия от предыдущего шага)
fig2 = go.Figure()
fig2.add_trace(go.Funnel(
    name='control',
    y=stages,
    x=funnel_df['control_from_prev'],
    textinfo="value+percent previous"
))
fig2.add_trace(go.Funnel(
    name='test',
    y=stages,
    x=funnel_df['test_from_prev'],
    textinfo="value+percent previous"
))
fig2.update_layout(title="Step-by-step воронка (% от предыдущего шага)")
fig2.show()
````

Сформулируем гипотезу для проведения теста

**Нулевая гипотеза (H₀):**
Для каждого уровня риска купленного актива доля пользователей, дошедших до второго депозита в тестовой группе, равна доле в контрольной группе.

**Альтернативная гипотеза (H₁):**
Для хотя бы одного уровня риска эта доля отличается между группами (двусторонняя проверка).

````python
# покупки и вторые депозиты
purchases_ctrl = abt[(abt['event_name']=='asset_purchase') & (abt['group']=='control')]
purchases_test = abt[(abt['event_name']=='asset_purchase') & (abt['group']=='test')]
deposits      = abt[abt['event_name']=='second_deposit']

results = []
for lvl in purchases_ctrl['risk_level'].dropna().unique():
    # контроль
    buyers_ctrl = purchases_ctrl[purchases_ctrl['risk_level']==lvl]['user_id'].unique()
    deps_ctrl   = deposits[deposits['user_id'].isin(buyers_ctrl)]['user_id'].unique()
    n_ctrl = len(buyers_ctrl)
    x_ctrl = len(deps_ctrl)
    # тест
    buyers_test = purchases_test[purchases_test['risk_level']==lvl]['user_id'].unique()
    deps_test   = deposits[deposits['user_id'].isin(buyers_test)]['user_id'].unique()
    n_test = len(buyers_test)
    x_test = len(deps_test)
    # Z-тест пропорций
    z, p = proportions_ztest([x_ctrl, x_test], [n_ctrl, n_test])
    results.append({
        'risk_level'   : lvl,
        'n_control'    : n_ctrl,
        'x_control'    : x_ctrl,
        'rate_control' : x_ctrl / n_ctrl,
        'n_test'       : n_test,
        'x_test'       : x_test,
        'rate_test'    : x_test / n_test,
        'p_value'      : p
    })

stats_df = pd.DataFrame(results).sort_values('risk_level')
stats_df[['risk_level','n_control','rate_control','n_test','rate_test','p_value']] \
       .style.format({
           'rate_control': '{:.1%}',
           'rate_test'   : '{:.1%}',
           'p_value'     : '{:.3f}'
       })
````

Сделайте промежуточный вывод. 


**Гипотезы для анализа метрик A/B-теста**

1. Ключевая метрика – средняя сумма всех депозитов на пользователя (включая нули).
   - H₀: средняя сумма депозитов в тестовой группе равна средней сумме депозитов в контрольной.
   - H₁: средняя сумма депозитов в тестовой группе отличается от средней суммы депозитов в контрольной.
2. Барьерная метрика – конверсия из регистрации в первый депозит.
   - H₀: доля пользователей, сделавших первый депозит, одинакова в тестовой и контрольной группах.
   - H₁: доля пользователей, сделавших первый депозит, в тестовой группе отличается от доли в контрольной.
3. Вспомогательная метрика 1 – конверсия из первого во второй депозит.
   - H₀: вероятность открытия второго депозита у пользователей контрольной и тестовой групп одинакова.
   - H₁: вероятность открытия второго депозита в тестовой группе отличается от таковой в контрольной.
4. Вспомогательная метрика 2 – средняя сумма депозитов среди пользователей, открывших хотя бы один депозит.
   - H₀: средний чек у активных депозиторoв в тестовой группе равен среднему чеку в контрольной.
   - H₁: средний чек у активных депозиторoв в тестовой группе отличается от среднего чека в контрольной.

````python
# Построение user-level таблицы
users = (
    abt[['user_id','group']]
       .drop_duplicates('user_id')
)

# суммарный депозит на пользователя
deposits = (
    abt[abt['event_name'].str.contains('_deposit')]
       .groupby('user_id')['amount']
       .sum()
       .reset_index(name='total_deposit')
)

# флаги первого и второго депозитов
first_dep = abt.loc[abt['event_name']=='first_deposit',  'user_id'].unique()
second_dep = abt.loc[abt['event_name']=='second_deposit', 'user_id'].unique()

users = (
    users
      .merge(deposits, on='user_id', how='left')
      .assign(
          total_deposit = lambda df: df['total_deposit'].fillna(0),
          has_first_dep  = lambda df: df['user_id'].isin(first_dep).astype(int),
          has_second_dep = lambda df: df['user_id'].isin(second_dep).astype(int)
      )
)

# Ключевая метрика: средний суммарный депозит ---
km = (
    users
      .groupby('group')['total_deposit']
      .agg(mean_deposit='mean', std='std', n='count')
      .reset_index()
)
ctrl_mean = km.loc[km.group=='control','mean_deposit'].iloc[0]
km['abs_diff'] = km['mean_deposit'] - ctrl_mean
km['rel_diff'] = km['abs_diff'] / ctrl_mean

# простой t-test (по умолчанию equal_var=True)
ctrl_vals = users.loc[users.group=='control','total_deposit']
test_vals = users.loc[users.group=='test',   'total_deposit']
t_stat, p_val = ttest_ind(ctrl_vals, test_vals)  

km['p_value'] = km['group'].map(lambda g: p_val if g=='test' else np.nan)

print("=== Ключевая метрика ===")
display(km.style.format({
    'mean_deposit':'{:.2f}',
    'abs_diff'    :'{:.2f}',
    'rel_diff'    :'{:.1%}',
    'p_value'     :'{:.3f}'
}))


# --- 3) Барьерная метрика: регистрация → первый депозит ---
bar = []
for grp in ['control','test']:
    grp_df = users[users.group==grp]
    n_all   = len(grp_df)
    x1      = grp_df.has_first_dep.sum()
    bar.append({'group':grp, 'conv1':x1/n_all, 'x1':x1, 'n_all':n_all})

bar = pd.DataFrame(bar)
z1, p1 = proportions_ztest(bar['x1'], bar['n_all'])
ctrl_conv1 = bar.loc[bar.group=='control','conv1'].iloc[0]
bar['abs_diff'] = bar.conv1 - ctrl_conv1
bar['rel_diff'] = bar['abs_diff'] / ctrl_conv1
bar['p_value']  = bar['group'].map(lambda g: p1 if g=='test' else np.nan)

print("\n=== Барьерная метрика ===")
display(bar.style.format({
    'conv1'   : '{:.1%}',
    'abs_diff': '{:.1%}',
    'rel_diff': '{:.1%}',
    'p_value' : '{:.3f}'
}))


# --- 4) Конверсия 1→2 депозит ---
supp1 = []
for grp in ['control','test']:
    grp_df = users[(users.group==grp)&(users.has_first_dep==1)]
    n1 = len(grp_df)
    x2 = grp_df.has_second_dep.sum()
    supp1.append({'group':grp, 'conv2': x2/n1 if n1>0 else np.nan,
                  'x2':x2, 'n1':n1})

supp1 = pd.DataFrame(supp1)
z2, p2 = proportions_ztest(supp1['x2'], supp1['n1'])
ctrl_conv2 = supp1.loc[supp1.group=='control','conv2'].iloc[0]
supp1['abs_diff'] = supp1.conv2 - ctrl_conv2
supp1['rel_diff'] = supp1['abs_diff'] / ctrl_conv2
supp1['p_value']  = supp1['group'].map(lambda g: p2 if g=='test' else np.nan)

print("\n=== Конверсия 1→2 депозит ===")
display(supp1.style.format({
    'conv2'   : '{:.1%}',
    'abs_diff': '{:.1%}',
    'rel_diff': '{:.1%}',
    'p_value' : '{:.3f}'
}))


# --- 5) Средний депозит среди активных пользователей ---
supp2 = (
    users[users.total_deposit>0]
      .groupby('group')['total_deposit']
      .agg(mean_active='mean', std='std', n='count')
      .reset_index()
)
ctrl_active = supp2.loc[supp2.group=='control','mean_active'].iloc[0]
ctrl_vals_act = users[(users.group=='control')&(users.total_deposit>0)]['total_deposit']
test_vals_act = users[(users.group=='test')   &(users.total_deposit>0)]['total_deposit']
t3, p3 = ttest_ind(ctrl_vals_act, test_vals_act)  

supp2['abs_diff'] = supp2.mean_active - ctrl_active
supp2['rel_diff'] = supp2['abs_diff'] / ctrl_active
supp2['p_value']  = supp2['group'].map(lambda g: p3 if g=='test' else np.nan)

print("\n=== Средний депозит среди активных ===")
display(supp2.style.format({
    'mean_active':'{:.2f}',
    'abs_diff'   :'{:.2f}',
    'rel_diff'   :'{:.1%}',
    'p_value'    :'{:.3f}'
}))
````
